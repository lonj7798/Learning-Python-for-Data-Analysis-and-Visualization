```python
import numpy as np
import pandas as pd
```

```python
dframe1 = pd.DataFrame({'key': ['x', 'z', 'y', 'z', 'x', 'x'], 'data_set_1': np.arange(6)})
dframe1
```

```python
dframe2 = pd.DataFrame({'key': ['q', 'y', 'z'], 'data_set_2': [1,2,3]})
dframe2
```

```python
pd.merge(dframe1, dframe2)
# automatically overlap the values that were matched
```

```python
pd.merge(dframe1, dframe2, on='key')
```

```python
pd.merge(dframe1, dframe2, on = 'key', how = 'left')
# match to left(dframe1)
```

```python
pd.merge(dframe1, dframe2, on = 'key', how = 'right')
# match to right(dframe2)
```

```python
pd.merge(dframe1, dframe2, on = 'key', how = 'outer')
# include both of key(dframe1, dframe2)
```

```python
dframe3 = pd.DataFrame({'key': list('xxxyzz'), 'data__set_3': range(6)})
dframe3
```

```python
dframe4 = pd.DataFrame({'key':list('yyxxz'), 'data_set_4': np.arange(5)})
dframe4
```

```python
pd.merge(dframe3, dframe4)
```

```python
df_left = pd.DataFrame({'key1': ['SF', 'SF', 'LA'],
                        'key2': ['one', 'two', 'one'],
                        'left_data':[10,20,30]})
df_left
```

```python
df_right = pd.DataFrame({'key1': ['SF', 'SF', 'LA', 'LA'],
                        'key2': ['one','one', 'one','two'],
                        'right_data':[40,50,60,70]})
df_left
```

```python
pd.merge(df_left, df_right, on = ['key1', 'key2'], how = 'outer')
```

```python
pd.merge(df_left, df_right, on = 'key1')
```

```python
pd.merge(df_left, df_right, on = 'key1', suffixes = ('_lefty', '_righty'))
```
