gold price VS oil price

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas_datareader as pdr
import datetime as datetime
```

```python
prices = pdr.get_data_yahoo(['CL=F', '1328.T'],
                            start = datetime.datetime(2018, 1, 1),
                            end = datetime.datetime(2020,4,22)
                           )['Adj Close']
```

```python
prices.plot()
```

```python
prices.columns = ['Oil', 'Gold']
```

```python
corr = prices.dropna().corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap = sns.diverging_palette(256,0,sep=80,n=7,as_cmap=True),annot = True, mask = mask)
```

```python
sns.jointplot('Gold','Oil',data=prices,kind='scatter')
```

```python
sns.distplot(prices['Gold'].dropna(),bins=100,color='purple')
```

```python
prices.Gold.plot()
```

```python
prices.Oil.plot()
```

```python
prices100 = (prices.Oil + 50) * 30
```

```python
prices100.plot(color='black')
prices.Gold.plot(color='orange')
plt.legend()
```

```python
pungsan = pdr.get_data_yahoo(['103140.KS'],
                            start = datetime.datetime(2018, 1, 1),
                            end = datetime.datetime(2020,4,22)
                           )['Adj Close']
pungsan.head()
```

```python
KOR = pdr.get_data_yahoo(['KRW=F'], start=datetime.datetime.today()-datetime.timedelta(days=1),
                         end=datetime.datetime.today()
                        )['Adj Close']
KOR.head()
```

```python
copper_kor = copper_kor*10
```

```python
df = pdr.get_data_yahoo(['103140.KS', 'HG=F'],
                            start = datetime.datetime(2015, 1, 1),
                            end = datetime.datetime(2020,4,22)
                           )['Adj Close']
df.columns=['Pungsan', 'Copper']
df.Copper = (df['Copper'] / (KOR.iloc[0,0]))*10
```

```python
df.plot()
```

```python
corr = df.dropna().corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap = sns.diverging_palette(256,0,sep=80,n=7,as_cmap=True),annot = True, mask = mask)
```

```python
sns.lmplot('Pungsan', 'Copper', df, markers='x')
```

```python
ma_day = ['10D','20D','50D']

for ma in ma_day:
    column_name_cu = "Copper for %says" %(str(ma))
    column_name_pung = "Pungsan for %says" %(str(ma))
    df[column_name_cu]=df['Copper'].rolling(ma).mean()
    df[column_name_pung]=df['Pungsan'].rolling(ma).mean()
```

```python
df.head()
```

```python
df[['Copper for 10Days','Copper for 20Days','Copper for 50Days']].plot(subplots=False,figsize=(10,4))
```

```python
df[['Pungsan for 10Days','Pungsan for 20Days','Pungsan for 50Days']].plot(subplots=False,figsize=(10,4))
```

```python
df['Copper for 50Days'].plot(figsize=(10,4))
df['Pungsan for 50Days'].plot(figsize=(10,4))
plt.legend()
```
