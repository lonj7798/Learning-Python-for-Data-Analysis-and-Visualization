```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
# 10 kinds of bins

# plt.hist: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color = 'indianred')
```

```python
plt.hist(dataset1, normed = True, color = 'indianred', alpha = 0.5, bins = 20)
plt.hist(dataset2, normed = True, alpha = 0.5, bins = 20)
```

```python
plt.hist(dataset1, density = True, color = 'indianred', alpha = 0.5, bins = 20)
plt.hist(dataset2, density = True, alpha = 0.5, bins = 20)

# use density instead of normed.
# density, normed: match the differnt number of array
```
