#### Box plot: https://en.wikipedia.org/wiki/Box_plot

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data = [data1, data2])
```

```python
sns.boxplot(data = [data1, data2], whis = np.inf)

# icludes the dot out of the range (data1)
```

```python
# sns.boxplot(data1, vertical = False)

# horizontal
sns.boxplot(data1, orient = 'h')
```

```python
# vertical
sns.boxplot(data1, orient = 'v')
```

```python
# information about curnel density

# Normal Distribution
data1 = stats.norm(0,5).rvs(100)

# Two gamma distribution. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50) - 1, 
                        -1 * stats.gamma(5).rvs(50)])

# Box plot both data1 and data2
sns.boxplot(data = [data1, data2], whis = np.inf)

# scipy.stats.norm: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
```

```python
sns.violinplot(data = [data1, data2])
```

```python
sns.violinplot(data2, bw = 0.01)
```

```python
sns.violinplot(data1)
```

```python
sns.violinplot(data1, inner = "point")

# sns.violinplot: https://seaborn.pydata.org/generated/seaborn.violinplot.html
```
