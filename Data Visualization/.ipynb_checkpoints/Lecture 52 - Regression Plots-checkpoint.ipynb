```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# tips = sns.load_dataset("tips")
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
sns.lmplot('total_bill', 'tip', tips)

# sns.lmplot: https://seaborn.pydata.org/generated/seaborn.lmplot.html
# blue linear line is linear regression line
```

```python
sns.lmplot('total_bill', 'tip', tips,
          scatter_kws = {'marker': 'o', 'color': 'indianred'},
          line_kws = {'linewidth': 1, 'color': 'blue'})
```

```python
# multipoe order

sns.lmplot('total_bill', 'tip', tips, order = 4,
          scatter_kws = {'marker': 'o', 'color': 'indianred'},
          line_kws = {'linewidth': 1, 'color': 'blue'})
```

```python
sns.lmplot('total_bill', 'tip', tips, fit_reg = False)
# only scatter plot
```

```python
tips.head()
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])
# tip percentage
```

```python
tips.head()
```

```python
sns.lmplot('size', 'tip_pect', tips)
```

#### Jitter?
https://en.wikipedia.org/wiki/Jitter

```python
sns.lmplot('size', 'tip_pect', tips, x_jitter = .1)
```

```python
# tendency of each bins

sns.lmplot('size', 'tip_pect', tips, x_estimator = np.mean)
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue = 'sex', markers = ['x', 'o'])
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue = 'day')
```

#### Local regression
https://en.wikipedia.org/wiki/Local_regression

```python
sns.lmplot('total_bill', 'tip_pect', tips, lowess = True,
          line_kws = {'color': 'orange'})
```

```python
sns.regplot('total_bill', 'tip_pect', tips)
```

```python
fig, (axis1, axis2) = plt.subplots(1,2,sharey = True)

sns.regplot('total_bill', 'tip_pect', tips, ax = axis1)
sns.violinplot(data = (tips['tip_pect'], tips['size']), ax = axis2)
```
