```python
nfl_frame = pd.read_clipboard()
nfl_frame

# ctl + C: copy on clipboard()
```

```python
nfl_frame = nfl_frame.rename(columns = {'1': 'Rank','Dallas Cowboys': 'Team','914': 'GP','520': 'Win','388': 'Lost','6': 'Tied','.572': 'Pct','1960':'First NFL Season','NFC East': 'Division'})
nfl_frame
```

```python
temp = pd.DataFrame({'Rank':['1'],'Team': ['Dallas Cowboys'],'GP': ['914'],'Win': ['520'],'Lost':[ '388'],'Tied':[ '6'],'Pct':[ '.572'],'First NFL Season': ['1960'],'Division': ['NFC East']})
nfl_frame = nfl_frame.append(temp)
nfl_frame
```

```python
nfl_frame.info()
```

```python
nfl_frame['Rank'] = nfl_frame['Rank'].astype(int)
nfl_frame = nfl_frame.sort_values(by = ['Rank'])
nfl_frame
```

```python
nfl_frame = nfl_frame.reset_index()
nfl_frame
```

```python
nfl_frame = nfl_frame.drop('index', axis = 1)
nfl_frame
```
